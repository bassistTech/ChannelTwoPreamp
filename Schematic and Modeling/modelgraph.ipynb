```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readfile(filename):
    s = open(filename).read()
    t = s.splitlines()
    f = list(map(float, [x.split('\t')[0] for x in t][1:]))
    g = list(map(float, [x.split('\t')[1].split(',')[0][1:-2] for x in t][1:]))
    return f, g

f, g = readfile('model1.txt')
plt.semilogx(f, g, label = '35 Hz')
f, g = readfile('model2.txt')
plt.semilogx(f, g, label = '70 Hz')
f, g = readfile('model3.txt')
plt.semilogx(f, g, label = '130 Hz')
plt.legend()
plt.title('HPF response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (dB)')
```

```python
r71 = 1000
r72 = 10000
alpha = np.linspace(0.01, 1, 100)
r72a = r72 * (1 - alpha)
r72b = r72 * alpha
gain = 20 * np.log10(r72b/(r72a + r71))
plt.figure(figsize = (4, 3))
plt.plot(alpha, gain)
plt.xlabel('knob position')
plt.ylabel('Gain (dB)')
plt.title('Gain vs knob position')
```
